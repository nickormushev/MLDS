```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import lemmagen.lemmatizer
from lemmagen.lemmatizer import Lemmatizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.manifold import TSNE
import textwrap
```

```python
df = pd.read_json('rtvslo_keywords.json')
```

I apply TF-IDF with a custom tokenizer so I can properly process keywords that are made up of 2 or words seperated by a space.

```python
concatenated_keywords = df['gpt_keywords'].map(lambda x: '$split$'.join(x).lower())
    
lemmatizer = Lemmatizer(dictionary=lemmagen.DICTIONARY_SLOVENE)

def comma_tokenizer(s):
    phrases = s.split('$split$')
    lemmatized_phrases = []
    for phrase in phrases:
        words = phrase.split()
        lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
        lemmatized_phrase = ' '.join(lemmatized_words)
        lemmatized_phrases.append(lemmatized_phrase)
    return lemmatized_phrases

vectorizer = TfidfVectorizer(min_df=20, tokenizer=comma_tokenizer, lowercase=False)

X = vectorizer.fit_transform(concatenated_keywords)
tfidf_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
```

I fit and run PCA with 3 components.

```python
pca = PCA(n_components=3)
X_pca = pca.fit_transform(tfidf_df)
```

```python
def biplot_3d(scores, loadings, features=None):
    fig = plt.figure(figsize=(7,7))
    ax = fig.add_subplot(111, projection='3d')
    xs = scores[:,0]
    ys = scores[:,1]
    zs = scores[:,2]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    scalez = 1.0/(zs.max() - zs.min())

    arrows = loadings.copy()  * 2.5

    ax.scatter(xs * scalex, ys * scaley, zs * scalez)

    max_indices = np.argpartition(arrows, -2, axis=0)[-2:]
    min_indices = np.argpartition(arrows, 2, axis=0)[:2]

    max_indices = max_indices.flatten()
    min_indices = min_indices.flatten()

    # features as arrows
    for i in set(max_indices.tolist() + min_indices.tolist()):
        ax.quiver(0, 0, 0, *arrows[i], color='g', alpha=1)
        ax.text(*(arrows[i] * 1.05), features[i], ha='center', va='center', 
                 bbox=dict(facecolor='white', alpha=1, edgecolor='none'), zorder=1000)


    ax.view_init(elev=0., azim=50)

# Assuming that your PCA transformed data is stored in X_pca and the PCA model in pca
biplot_3d(X_pca[:, :3],pca.components_[:3].T, features=tfidf_df.columns)
plt.show()
```

In the plot we see which words most influneced the first 3 principle components while they were trying to preserve variance. We can see one is related to sports with keywords like zmaga and tekma which are related to sports. Rusija, Ukrajna identifying topics related to war. Umetnost and razstava identify sections related to art.

# Task 2 T-SNE

For this part first I had the idea to run a grid search on PCA and K-means to find the best combination of n_components and n_clusters to later use to cluster the data. For a metric I used the silhoutte score. This lead to a few issues. First less components lead to a better silhoutte score since less features are easier to cluster which makes me believe this is not an optimal way to chose the components at least but only the number of clusters. 

Before this I had actually tried to use a scree plot for n_components but the results seen below are not very infromative since the first components preserved a very small amount of variance(around 1%). I need approximately 1500 components to preserve the variance to around 80% which does not seem optimal to me since we want a bigger reduction in features. As a result I ended up testing a bit more and using 11 as the number of components and estimating that 13 is the best number of clusters using the grid search.

```python
pca = PCA(n_components=2000)

pca.fit(tfidf_df)

explained_variance = pca.explained_variance_ratio_

cumulative_explained_variance = np.cumsum(explained_variance)

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, linewidth=2, label='Individual')
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, linewidth=2, label='Cumulative')
plt.title('Scree Plot')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

Below is my somewhat failed grid search

```python
n_components_range = [11]
n_clusters_range = range(8, 50)

best_score = -1
scores = {}

for n_components in n_components_range:
    print(f'Number of PCA components: {n_components}')
    pca = PCA(n_components=n_components)
    X_pca = pca.fit_transform(X)

    for n_clusters in n_clusters_range:
        print(f'Number of clusters: {n_clusters}')
        kmeans = KMeans(n_clusters=n_clusters, random_state=0)
        labels = kmeans.fit_predict(X_pca)

        score = silhouette_score(X_pca, labels)
        scores[n_clusters] = score

        if score > best_score:
            best_score = score
            best_n_components = n_components
            best_n_clusters = n_clusters

print(f'Best number of PCA components: {best_n_components}')
print(f'Best number of KMeans clusters: {best_n_clusters}')
print(f'Best silhouette score: {best_score}')
```

Using the plot below I chose 13 cluster as the silhoutte score there was highest

```python
plt.figure(figsize=(10, 6))
plt.plot(list(scores.keys()), list(scores.values()))
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
max_score_clusters = max(scores, key=scores.get)

plt.axvline(x=max_score_clusters, color='r', linestyle='--')

xticks = plt.xticks()[0]
xticks = [tick for tick in xticks if tick not in [0, 60]]

plt.xticks(list(xticks) + [max_score_clusters])

plt.show()
```

After finding the best value for number of clusters I run PCA and k-means again

```python
pca = PCA(n_components=11)
X = pca.fit_transform(tfidf_df)

kmeans = KMeans(n_clusters=best_n_clusters, random_state=0)
kmeans.fit(X)
labels = kmeans.labels_
```

Because of the many datapoints we have I deceded to increase perplexity telling T-SNE how big the regions it should look are. Basically giving a guess that each point should expect to have around 50 close neighbors since based on what I have seen regions contain a lot of samples. This value was also chosen based on the recommendation of the T-SNE paper that we should use a perplexity between 5 and 50. I also increased the max_iteration to make sure it converges. I saw that  the implementation of scikit-learn checks for early stopping so I believe this should work better.

```python
tsne = TSNE(n_components=2, random_state=0, perplexity=50, max_iter=10000)

# Fit and transform the data
X_tsne = tsne.fit_transform(X)
```

In the following section I extract the top 3 keyowrds based on sum of TF-IDF scores and use them to lable the clusters. I need more than one word cause it is common for sports sections to have for example zmaga/victory as a most keyword so two clusters are identified by the same keywords which reduces interpretability.

```python
X_inv = pca.inverse_transform(X)

keywords = tfidf_df.columns
cluster_keywords = []
for i in range(kmeans.n_clusters):
    cluster = np.where(labels == i)
    cluster_data = X_inv[cluster] 
    keyword_sums = cluster_data.sum(axis=0)
    top_3_keyword_indices = np.argsort(keyword_sums)[-3:]
    top_3_keywords = keywords[top_3_keyword_indices]
    concatenated_keywords = ', '.join(top_3_keywords)
    cluster_keywords.append(concatenated_keywords)
```

```python
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab20')

for i, keyword in enumerate(cluster_keywords):
    wrapped_keyword = textwrap.fill(keyword, width=10)  
    centroid = np.mean(X_tsne[labels == i], axis=0)

    plt.annotate(wrapped_keyword,
                  (centroid[0], centroid[1]),
                  fontsize=8,
                  bbox=dict(boxstyle='round,pad=0.5', fc='grey', alpha=0.5))

plt.show()
```
